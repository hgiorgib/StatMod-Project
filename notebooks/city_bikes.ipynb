Send a request to CityBikes for the city of your choice. 

```python
import requests
import pandas as pd
import json
```

```python
#Request JSON files from Citybik API for the city of Toronto, ON, in Canada
Toronto = requests.get('http://api.citybik.es/v2/networks/bixi-toronto').json() 
```

Parse through the response to get the details you want for the bike stations in that city (latitude, longitude, number of bikes). 

```python
#Parse JSON structure of Toronto file
print(json.dumps(Toronto,indent=4))
```
